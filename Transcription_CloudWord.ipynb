```python
from PyPDF2 import PdfReader
from textblob import TextBlob
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt

# File path
file_path = "//Users//myueta//Documents//MyU24//DataAnalytics//Projects//Transcription_CloudWord//META-Q4-2023-Earnings-Call-Transcript.pdf"

# Read the PDF file and extract text
reader = PdfReader(file_path)
text = ""
for page in reader.pages:
    text += page.extract_text()

# Delete words that do not contribute to the analysis
del_words = set(STOPWORDS)
del_words.update(["will", "year", "really", "going", "across"])

# Generate word cloud
word_cloud = WordCloud(stopwords=del_words, background_color='black', 
                       width=1920, height=1080).generate(text)

# Display the word cloud
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis("off")
plt.show()

# Perform sentiment analysis using TextBlob
blob = TextBlob(text)
polarity = blob.sentiment.polarity

# Display the sentiment polarity
print("Sentiment Polarity:", polarity)


```
